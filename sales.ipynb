```python
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)
    
cat_cols = df.select_dtypes(include=["object"]).columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)
df_encoded.head()
```

```python
scaler = StandardScaler()

num_cols_encoded = df_encoded.select_dtypes(include=["int64", "float64"]).columns
df_encoded[num_cols_encoded] = scaler.fit_transform(df_encoded[num_cols_encoded])

df_encoded.head()
```

```python
scaler = MinMaxScaler()

df_encoded[num_cols_encoded] = scaler.fit_transform(df_encoded[num_cols_encoded])
```

```python
normalizer = Normalizer()

df_encoded[num_cols_encoded] = normalizer.fit_transform(df_encoded[num_cols_encoded])
```

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
df["DayOfWeek"] = df["Date"].dt.dayofweek
df["Quarter"] = df["Date"].dt.quarter
df["IsWeekend"] = df["DayOfWeek"].apply(lambda x: 1 if x >= 5 else 0)
```

```python
df["Revenue"] = df["Unit_Price"] * df["Order_Quantity"]
```

```python
df["Avg_Order_Value"] = df["Revenue"] / (df["Order_Quantity"] + 1)
```

```python
df["Price_per_Unit"] = df["Revenue"] / (df["Order_Quantity"] + 1)
```

```python
fig = px.histogram(df, x="Revenue", nbins=50,
                   title="Revenue Distribution",
                   color_discrete_sequence=["royalblue"])
fig.show()
```

```python
category_rev = df.groupby("Product_Category")["Revenue"].sum().reset_index()

fig = px.bar(category_rev,
             x="Product_Category",
             y="Revenue",
             title="Total Revenue by Product Category",
             color="Revenue")
fig.show()
```

```python
year_rev = df.groupby("Year")["Revenue"].sum().reset_index()

fig = px.line(year_rev,
              x="Year",
              y="Revenue",
              markers=True,
              title="Revenue Over Years")
fig.show()
```

```python
fig = px.scatter(df,
                 x="Revenue",
                 y="Profit",
                 color="Product_Category",
                 title="Profit vs Revenue")
fig.show()
```
