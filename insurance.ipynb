```python
df["log_charges"] = np.log1p(df["charges"])
```

```python
fig1 = px.histogram(df, x="charges", title="Original Charges Distribution", nbins=40)
fig1.show()

fig2 = px.histogram(df, x="log_charges", title="Log Transformed Charges Distribution", nbins=40)
fig2.show()
```

```python
y = df["log_charges"]
```

```python
df_encoded = pd.get_dummies(df, columns=["sex", "smoker", "region"], drop_first=True)   
```

```python
x = df_encoded.drop(columns=["charges", "log_charges"])
y = df_encoded["log_charges"]
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
print("Data split into training and testing sets successfully.")
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)        
print("y_train shape:", y_train.shape)  
print("y_test shape:", y_test.shape)
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
r2_score = model.score(x_test, y_test)
print(f"R² Score: {r2_score:.4f}")
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.4f}") 
rmse = root_mean_squared_error(y_test, y_pred)
print(f"Root Mean Squared Error: {rmse:.4f}")
```
